```python
x = np.linspace(-5,5,50)
f = lambda x: x**2
plt.plot(x,f(x))

x0 = 4                     
l  = 1e-5
GD = lambda x0 : x0 - 0.1 * (2*x0)
for x in range(1,100):
    x1 = GD(x0)
    if((x0-x1)<l):       # here it check x0-x1 is less than 0ne or not 
        print(x1)
        break
    plt.scatter(x1,f(x1))  
    x0 = x1
plt.title('covergence of x^2')          # keeping the heading
plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
plt.ylabel('Y-axis', fontsize = 15)         # label the y axis    
plt.show

  
```

```python
def GD_Function(f,df,ip,lr,l):
    x = np.linspace(-5,5,100)     # we get 100 point between -5 and 5
    plt.plot(x,f(x))              # here it plat for (x,f(x))
    
    GD = lambda x0,lr : x0 - (lr) * df(x0)
    for x in range(100):
        x1 = GD(ip,lr)
        plt.scatter(x1,f(x1)) 
        
        if(np.abs(ip-x1)<l):
            print(x1)
            break
         
        ip = x1
    plt.title('covergence of x^2')          # keeping the heading
    plt.xlabel('X-axis', fontsize = 15)         # label the x axis 
    plt.ylabel('Y-axis', fontsize = 15)         # label the y axis        
    plt.show
```

```python
f = lambda x: x**2        # define a f and df out side we send into function
df = lambda x: 2*x
GD_Function(f,df,4,0.1,1e-5)
```
