```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Given f(ùë•1, ùë•2) = 4ùë•1 + 6ùë•2 ‚àí 2ùë•12 ‚àí 2ùë•1ùë•2 ‚àí 2ùë•2


```python
fig = plt.figure(figsize=plt.figaspect(0.5))      # set up a figure as wide as it is tall
ax = fig.add_subplot(1,2,1,projection = '3d')     # set up the axes for the first plot
x1= np.arange(-20,20,1)
x2= np.arange(-20,20,1)
x1,x2 = np.meshgrid(x1,x2)
fun = lambda x1,x2: 4*(x1) + 6*(x2) - 2*(x1)**2 - 2*(x2)**2 -2*(x2)  # this is given function
ax.scatter3D(x1,x2,fun(x1,x2),color = 'b') 
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
